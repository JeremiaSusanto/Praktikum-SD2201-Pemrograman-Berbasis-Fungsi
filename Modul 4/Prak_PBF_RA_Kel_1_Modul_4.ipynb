```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

def correlation_matrix_generator():
    def generate(data):
        # Membuat DataFrame dari data yang diberikan
        df = pd.DataFrame(data)

        # Menghitung matriks korelasi dari DataFrame
        corr_matrix = df.corr()

        # Menampilkan heatmap dari matriks korelasi dengan nilai-nilai di dalamnya
        sns.heatmap(corr_matrix, annot=True)
        plt.show()

    return generate

# Mendefinisikan fungsi generator matriks korelasi
generate_corr_matrix = correlation_matrix_generator()

# Contoh penggunaan dengan data dummy
data_dummy = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6]
}

# Menghasilkan matriks korelasi untuk data dummy
generate_corr_matrix(data_dummy)

# Mendefinisikan dataset baru
data_baru = {
    'X': [3, 6, 9, 12, 15],
    'Y': [10, 8, 6, 4, 2],
    'Z': [7, 5, 3, 1, 0]
}

# Menghasilkan matriks korelasi untuk data baru
generate_corr_matrix(data_baru)
```
